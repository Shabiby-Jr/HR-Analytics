```python
def convert_comsize_to_num(x):
    tokens = x.split("-")
    if len(tokens) == 2 :
        return(float(tokens[0]) + float(tokens[1]))/2
    else:
        return x
    try:
        return float(x)
    except:
        return None
```

```python
data.company_size = data.company_size.apply(convert_comsize_to_num)
data.head(10)
```

```python
def convert_company(x):
    if x < 100:
        return "small_company"
    elif x < 1000:
        return "medium_company"
    elif x < 5000:
        return "large_company"
    else:
        return "multinational_company"
    
    
data.company_size = data.company_size.apply(convert_company)
data.tail(10)
    
    
    
```

```python
company = pd.get_dummies(data.company_size)
company.drop("small_company", axis="columns", inplace = True)

data = pd.concat([data,company], axis=1)

data.drop("company_size", axis=1, inplace = True)
data.tail(10)
```

```python
data.info()
```

```python
X = data.drop('target', axis='columns')
X.head()
```

```python
y = data.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 10)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators=50, random_state=0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(roc_auc_score(y_test, y_pred))
```
